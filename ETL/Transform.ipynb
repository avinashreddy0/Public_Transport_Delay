```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r'C:\Users\indur\OneDrive\Desktop\power bi projects\Public_Transport_Delay\ETL\public_transport_delay_messy_45000.csv')
df
```

```python
df.columns = df.columns.str.upper()
df
```

```python
#about data before the data cleaning 
print(df.describe())
print(df.info())
print(df.isna().sum())
print(df.shape)
print(df.sample(10))
print(df.head())
print(df.tail())
df.duplicated().sum()
print(df.dtypes)
```

```python
print('---------------------------------------------CLEANING--------------------------------------------------')
```

```python
# trip_id column

df['TRIP_ID'] = range(1,len(df)+1)
df['TRIP_ID'] = 'TR' + df['TRIP_ID'].astype(str)
df
```

```python
#cleaning city column

df['CITY'] = df['CITY'].str.replace(r'[^A-Za-z]','',regex=True)
df['CITY'] =df['CITY'].replace({'Hydderbad':'Hyderabad','Dlhi':'Delhi','Chenni':'Chennai','Mumbi':'Mumbai','Bnglre':'Bangalore'},regex=False).astype('category')
df['CITY'] = df['CITY'].str.strip().str.title()
df
```

```python

#cleaning route_id

df['ROUTE_ID'] =df['ROUTE_ID'].str.replace(r'[^A-Za-z0-9_]','',regex=True)
df['ROUTE_ID'] = df['ROUTE_ID'].str.replace('Bu','bus')
df['ROUTE_ID'] = df['ROUTE_ID'].str.replace('0','o')
df['ROUTE_ID'] = df['ROUTE_ID'].str.replace('Trin','Train')
df['ROUTE_ID'] = df['ROUTE_ID'].str.strip().str.title()
df
```

```python
#cleaning hour column 

df['HOUR'] =df['HOUR'].str.extract(r'(\d+)')
df
```

```python
df['RAINFALL_MM'] = df['RAINFALL_MM'].astype(str)
df['RAINFALL_MM'] = df['RAINFALL_MM'].str.replace(r'[^0-9.]','',regex=True)
df['RAINFALL_MM'] = df['RAINFALL_MM'].str.replace(r'\.+','.',regex=True)
df
```

```python
#cleaning day_of_week

df['DAY_OF_WEEK'] = df['DAY_OF_WEEK'].str.replace(r'[^A-Za-z]','',regex=True)
df['DAY_OF_WEEK'] = df['DAY_OF_WEEK'].replace({'Saturday':'saturday','Frday':'friday','Sun':'sunday','Mon':'monday','Wd':'wednesday'})
df['DAY_OF_WEEK'] = df['DAY_OF_WEEK'].str.strip().str.title()
df
```

```python
#cleaning temperature column

df['TEMPERATURE'] = df['TEMPERATURE'].str.replace(r'[^0-9.]','',regex=True)
df['TEMPERATURE'] = df['TEMPERATURE'].str.replace(r'\.+','.',regex=True)
df
```

```python
#cleaning humidity

df['HUMIDITY'] = df['HUMIDITY'].str.replace(r'[^0-9.]','',regex=True)
df['HUMIDITY'] = df['HUMIDITY'].str.replace(r'\.+','.',regex=True)
df
```

```python
#cleaning wind_speed

df['WIND_SPEED'] = df['WIND_SPEED'].str.replace(r'[^A-Za-z0-9/]','',regex=True)
df
```

```python
#CLEANING RUSH_HOUR

df['RUSH_HOUR'] = df['RUSH_HOUR'].str.replace(r'[^A-Za-z0-9]','',regex=True)
df['RUSH_HOUR'] = df['RUSH_HOUR'].replace({'0':'no','1':'yes','True':'yes','False':'no'})
df['RUSH_HOUR'] = df['RUSH_HOUR'].str.strip().str.title()

df
```

```python
#cleaning TRAFFIC_LEVEL

df['TRAFFIC_LEVEL'] = df['TRAFFIC_LEVEL'].str.replace('0','o')
df['TRAFFIC_LEVEL'] = df['TRAFFIC_LEVEL'].str.replace('@','a')
df['TRAFFIC_LEVEL'] = df['TRAFFIC_LEVEL'].str.replace('!','i')
df['TRAFFIC_LEVEL'] = df['TRAFFIC_LEVEL'].str.strip().str.title()
df
```

```python
#cleaning EVENT_TODAY

df['EVENT_TODAY'] = df['EVENT_TODAY'].str.replace(r'[^A-Za-z]','',regex=True)
df['EVENT_TODAY'] = df['EVENT_TODAY'].replace({'Y':'Yes','N':'No'}).str.strip().str.title()
df
```

```python
#cleaning EVENT_TYPE

df['EVENT_TYPE'] = df['EVENT_TYPE'].str.replace('0','o')
df['EVENT_TYPE'] = df['EVENT_TYPE'].str.replace('@','a')
df['EVENT_TYPE'] = df['EVENT_TYPE'].str.replace('3','e')
df['EVENT_TYPE'] = df['EVENT_TYPE'].str.replace(r'[^A-Za-z_]','',regex=True)
df['EVENT_TYPE'] = df['EVENT_TYPE'].str.replace('Concertaa','Concert')
df
```

```python
#cleaning IS_WEEKEND

df['IS_WEEKEND'] = df['IS_WEEKEND'].str.replace(r'[^A-Za-z0-9]','',regex=True)
df['IS_WEEKEND'] = df['IS_WEEKEND'].replace({'0':'false','1':'true','Flse':'False'}).str.strip().str.title()
df['IS_WEEKEND'] = df['IS_WEEKEND'].replace({'False':'NO','True':'YES'})
df
```

```python
#cleaning VEHICLE_TYPE

df['VEHICLE_TYPE'] = df['VEHICLE_TYPE'].str.replace('0','o')
df['VEHICLE_TYPE'] = df['VEHICLE_TYPE'].str.replace('$','s')
df['VEHICLE_TYPE'] = df['VEHICLE_TYPE'].str.replace('@','a')
df['VEHICLE_TYPE'] = df['VEHICLE_TYPE'].str.replace(r'[^A-Za-z]','',regex=True)
df['VEHICLE_TYPE'] = df['VEHICLE_TYPE'].replace('BUsa','bus').str.title()
df
```

```python
#cleaning DELAY_MINUTES

df['DELAY_MINUTES'] = df['DELAY_MINUTES'].replace(r'[^0-9]','',regex=True)
df['DELAY_MINUTES'] = df['DELAY_MINUTES'].astype(int)
df
```

```python
print(df.describe())
print(df.info())
print(df.dtypes)
print(df.isnull().sum())
print(df.head())
print(df.tail())
```

```python
#saving data cleaning with csv and excel files

try:
    df.to_csv('public_transport_delay_cleaned.csv')
    df.to_excel('public_transport_delay_cleaned.xlsx')
except FileExistsError:
    print('please check')
finally:
    print('successfully we saved csv and excel files..............')
```
