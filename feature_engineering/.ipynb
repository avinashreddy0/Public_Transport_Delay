```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r'C:\Users\indur\OneDrive\Desktop\power bi projects\Public_Transport_Delay\public_transport_delay_cleaned.csv')
```

```python
df['WIND_SPEED'] = df['WIND_SPEED'].str.replace(r'[^0-9.]','',regex=True)
df['WIND_SPEED'] = df['WIND_SPEED'].astype(float)
```

```python
df['EVENT_TYPE'] = df['EVENT_TYPE'].fillna('IPL_Match')
df
```

```python
df['DELAY_MINUTES'] = df['DELAY_MINUTES'].replace(r'[^0-9]','',regex=True)
df['DELAY_MINUTES'] = df['DELAY_MINUTES'].astype(int)
df
```

```python
# aggregation feature engineering

df['AVG_DELAY_PER_CITY']  = df.groupby('CITY')['DELAY_MINUTES'].transform('mean')
df
```

```python
# aggregation feature engineering

df['AVG_DELAY_PER_CITY']  = df.groupby('CITY')['DELAY_MINUTES'].transform('mean')
df
```

```python
df["RUSH_HOUR"] = df["RUSH_HOUR"].map({"Yes":1,"No":0,"True":1,"False":0})
df["EVENT_TODAY"] = df["EVENT_TODAY"].map({"Yes":1,"No":0})
df["IS_WEEKEND"] = df["IS_WEEKEND"].map({"YES":1,"NO":0})
df
```

```python
df["TRAFFIC_LEVEL"] = df["TRAFFIC_LEVEL"].str.replace(r"[^A-Za-z_]", "", regex=True)
df
```

```python
df['CITY'] = df['CITY'].map({'Hyderabad':1,'Bangalore':2,'Chennai':3,'Mumbai':4,'Delhi':5})
df['CITY'] = df['CITY'].fillna(df['CITY'].median())
df
```

```python
df['AVG_DELAY_PER_TRAFFIC_LEVEL'] = df.groupby('TRAFFIC_LEVEL')['DELAY_MINUTES'].transform('mean')
df
```

```python
df['TRAFFIC_LEVEL'] = df['TRAFFIC_LEVEL'].map({'Low_Traffic':1,'Medium_Traffic':2,'High_Traffic':3})
df['TRAFFIC_LEVEL'] =df['TRAFFIC_LEVEL'].fillna(df['TRAFFIC_LEVEL'].median())
df
```

```python
df['ROUTE_ID'] = df['ROUTE_ID'].map({'Busss88':1,'Busss55':2,'Local_Train':3,'Metrored':4,'Metrored':5})
df['ROUTE_ID'] = df['ROUTE_ID'].fillna(df['ROUTE_ID'].median())
df




df['EVENT_TYPE'] = df['EVENT_TYPE'].map({'IPL_Match':1,'Cricket_Match':2,'Concert':3})
df['EVENT_TYPE'] = df['EVENT_TYPE'].fillna(df['EVENT_TYPE'].median())
df

df['VEHICLE_TYPE'] = df['VEHICLE_TYPE'].map({'Metro':1,'Bus':2,'Train':3})
df['VEHICLE_TYPE'] = df['VEHICLE_TYPE'].fillna(df['VEHICLE_TYPE'].median())
df
```

```python
# feature engineering 

df['DAY_OF_WEEK'] = df['DAY_OF_WEEK'].map({'Monday':1,'Tuesday':2,'Wednesday':3,'Thursday':4,'Friday':5,'Saturday':6,'Sunday':7})
df
```

```python

df['HUMIDITY_TEMP_RATIO'] = df['HUMIDITY'] / df['TEMPERATURE']
```

```python
df.to_csv('cleaned_data.csv',index=False)
```
